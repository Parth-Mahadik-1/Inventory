```python
import pandas as pd
mail = pd.read_csv("/content/drive/MyDrive/colab collection/spam.csv")
mail.head()
```

```python
dummys = pd.get_dummies(mail.Category)
dummys.head()
```

```python
mail = pd.concat([mail,dummys],axis="columns")
mail = mail.drop("Category",axis="columns")


```

```python
mail.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(mail.Message,mail.spam,test_size=0.25)
```

```python
v = CountVectorizer()
X_train_count = v.fit_transform(X_train.values)
X_train_count.toarray()[:3]
```

```python
from sklearn.naive_bayes import MultinomialNB
model = MultinomialNB()
model.fit(X_train_count,y_train)
```

```python
emails = [
    'Hey mohan, can we get together to watch footbal game tomorrow?',
    'Upto 20% discount on parking, exclusive offer just for you. Dont miss this reward!'
]
emails_count = v.transform(emails)
model.predict(emails_count)
```

```python
X_test_count = v.transform(X_test)
model.score(X_test_count, y_test)
```

with pipline method

```python
from sklearn.pipeline import Pipeline
a = Pipeline([("vectorizer",CountVectorizer()),
              ("nb",MultinomialNB())])
```

```python
a.fit(X_train,y_train)
```
