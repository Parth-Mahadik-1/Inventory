```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model

```

**Reading csv name called m1.csv, and creating its scatter plot **

```python
df = pd.read_csv("/content/drive/MyDrive/colab collection/m1.csv")
plt.scatter(df["area"],df['price'],color="red",marker="+")
plt.xlabel("area[sqft]")
plt.ylabel("price[us$]")
plt.title(" house pridiction ")
```

**here we create an object which use to feed our data.**

```python
reg=linear_model.LinearRegression()
```

**here is use of reg object**

```python
reg.fit(df[["area"]],df["price"])
```

this is our **output**

```python
reg.predict([[850]])
```
