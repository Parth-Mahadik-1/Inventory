```python
import pandas as pd

titanic = pd.read_csv("/content/drive/MyDrive/colab collection/titanic.csv")
```

```python
titanic.head(3)
```

```python
titanic.drop(["PassengerId","Name","SibSp","Parch","Ticket","Cabin","Embarked"],axis="columns", inplace=True)
```

```python
y = titanic.Survived
```

```python
x = titanic.drop("Survived",axis="columns")
x.head(3)
```

```python
dummy = pd.get_dummies(x.Sex)
dummy.head(3)
```

```python
x = pd.concat([x,dummy],axis="columns")
```

```python
x.Age = x.Age.fillna(x.Age.mean())
```

```python
x.head(15)
```

```python
x = x.drop("Sex",axis="columns")
x.head()
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
nb = GaussianNB()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

```python
nb.fit(x_train,y_train)
```

```python
nb.score(x_test,y_test)
```
