```python
import pandas as pd
from sklearn import linear_model
```

```python
df = pd.read_csv("/content/drive/MyDrive/colab collection/m4.csv")
df
```

```python
a =int( df['bedroom'].median())
df['bedroom']=df['bedroom'].fillna(a)
df

```

```python
reg = linear_model.LinearRegression()
```

```python
reg.fit(df[["area","bedroom","age"]],df['price'])
```

```python
reg.predict([[3200,4,15]])
```

```python
reg.predict([[3000,3,40]])
```

```python
reg.predict([[2500,4,5]])
```
