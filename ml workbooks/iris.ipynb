```python
from sklearn import datasets
import pandas as pd
```

```python
df = datasets.load_iris()
```

```python
dir(df)
df.feature_names
```

```python
df_dataset=pd.DataFrame(df.data,columns=df.feature_names)
df_dataset["target"]=df.target
df_dataset
df_dataset["names"] = df_dataset.target.apply(lambda x:df.target_names[x])
df_dataset


```

```python
x = df_dataset.drop(columns=["target", "names"])
x
y = df_dataset.target
y
```

```python
from sklearn.model_selection import  train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,train_size=0.9)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
```

```python
model .fit(x_train,y_train)
```

```python
model.predict([[5.9,3.0,5.1,1.8]])
```
