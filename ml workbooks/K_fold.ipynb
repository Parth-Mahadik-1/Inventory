```python
from sklearn import datasets
from sklearn.model_selection import train_test_split
import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
iris = datasets.load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df["target"] = iris.target
df.head()
```

```python
x = iris.data
y = iris.target

from sklearn.model_selection import StratifiedKFold
folds = StratifiedKFold(n_splits=3)
```

```python
def getscore(model,x_train,x_test,y_train,y_test):
  model.fit(x_train,y_train)
  return model.score(x_test,y_test)
```

```python
l_score=[]
svm_score=[]
rf_score=[]
```

```python


for train_index, test_index in folds.split(x,y):
  x_train,x_test,y_train,y_test=x[train_index],x[test_index],y[train_index],y[test_index]
  l_score.append(getscore(LogisticRegression(max_iter=1000),x_train,x_test,y_train,y_test))
  svm_score.append(getscore(SVC(),x_train,x_test,y_train,y_test))
  rf_score.append(getscore(RandomForestClassifier(n_estimators=50),x_train,x_test,y_train,y_test))
```

```python
from sklearn.model_selection import cross_val_score
cross_val_score(LogisticRegression(max_iter=1000), x, y,cv=3)
```

```python
cross_val_score(SVC(), x, y,cv=3)
```

```python
cross_val_score(RandomForestClassifier(n_estimators=50), x, y,cv=3)
```
