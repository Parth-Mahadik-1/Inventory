#sigmoid fucntion


```python
import math
def sigmoid(x):
  return 1 / (1 + math.exp(-x))
```

```python
sigmoid(100)
```

```python
sigmoid(-2)
```

#tanh fucntion

```python
def tanh(x):
  return (math.exp(x) - math.exp(-x)) /(math.exp(x)+math.exp(-x))
```

```python
tanh(100)
```

```python
tanh(-2)
```

#Relu fucntion

```python
def relu(x):
  return max(0,x)
```

```python
relu(100)
```

```python
relu(-2)
```

#leaky relu fucntion


```python
def leaky_relu(x):
  return max(0.1*x,x)
```

```python
leaky_relu(100)
```

```python
leaky_relu(-2)
```

#LOSS Functions

1) MAE - mean abosulute error


```python
y_predicted = np.array([1,1,0,0,1])
y_true = np.array([0.30,0.7,1,0,0.5])
def mae(y_true , y_predicted):
  tot_error = 0
  for yp,yt in zip(y_predicted,y_true):
    tot_error = y_true-y_predicted
    mae = np.mean(np.abs(tot_error))
  print("total error:",tot_error)
  print("mae:",mae)
```

```python
mae(y_predicted,y_true)
```

2)MSE - Mean Squared Error


```python
y_predicted = np.array([1,1,0,0,1])
y_true = np.array([0.30,0.7,1,0,0.5])
def mse(y_true , y_predicted):
  squ_error = 0
  for yt,yp in zip(y_true,y_predicted):
    squ_error += (yt-yp)**2
    mse_value = squ_error / len(y_true)
  print("total error",squ_error)
  print("mse",mse_value)
```

```python
mse(y_true , y_predicted)
```

```python
np.mean(((y_true-y_predicted)**2))
```

3) log loss  or binary cross entropy

```python
epsilon = 1e-15
```

```python
y_predicted = np.array([1,1,0,0,1])
y_true = np.array([0.30,0.7,1,0,0.5])
```

```python
y_predicted_new = [ max(i,epsilon) for i in y_predicted]
y_predicted_new
```

```python
y_predicted_new = [ min(i,1-epsilon) for i in y_predicted_new]
y_predicted_new
np.log(y_predicted_new)
```

```python
def log_loss(y_true, y_predicted):
    epsilon = 1e-15
    y_predicted_new = [min(i, 1 - epsilon) for i in y_predicted]
    y_predicted_new = [max(i, epsilon) for i in y_predicted_new]
    y_predicted_new = np.array(y_predicted_new)
    return -np.mean(y_true * np.log(y_predicted_new) + (1 - y_true) * np.log(1 - y_predicted_new))


```

```python

log_loss(y_true,y_predicted)
```
