#Implement Gradient Descent For Neural Network (or Logistic Regression)

```python
import pandas as pd
import tensorflow as tf
from tensorflow import keras
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import numpy as np

```

```python
data = pd.read_csv('/content/drive/MyDrive/DL notebooks/titanic.csv')

```

## scaling data and create a fetures and targets variables

```python
x = data[['Pclass', 'Age', 'Sex']].copy()
y = data[['Survived']].values.flatten()

x['n_age'] = x['Age'].fillna(x['Age'].mean())
x['N_Sex'] = x['Sex'].apply(lambda x: 1 if x == 'male' else 0)
x.drop(["Sex", "Age"], axis=1, inplace=True)

scaler = StandardScaler()
x = scaler.fit_transform(x)

```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.1, random_state=42)
```

##training model using logistic regresion(sigmoid)


```python

model = keras.Sequential([
    keras.layers.Dense(1, input_shape=(3,), activation='sigmoid')
])

model.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy']
)
model.fit(x_train,y_train,epochs=100)
```

##Watching loss which near to in between  4.3 to 4.80

```python
model.evaluate(x_test,y_test)
```

```python
def log_loss(y_true, y_predicted):
    epsilon = 1e-15
    y_predicted_new = np.clip(y_predicted, epsilon, 1 - epsilon)
    return -np.mean(y_true * np.log(y_predicted_new) + (1 - y_true) * np.log(1 - y_predicted_new))
```

#Gradieant decent from python scratch

```python
def gradient_Decent(Pclass,Age,Sex,y_true, epochs, loss_threshold):
  w1 = w2 = w3 = 1
  bias = 0
  rate = 0.5
  n = len(Pclass)
  for i in range(epochs):
     weighted_sum = w1*Pclass + w2*Age + w3*Sex + bias
     y_predicted = sigmoid_numpy(weighted_sum)
     loss = log_loss(y_true, y_predicted)
     w1d = (1 / n) * np.dot(Pclass, (y_predicted - y_true))
     w2d = (1 / n) * np.dot(Age, (y_predicted - y_true))
     w3d = (1 / n) * np.dot(Sex, (y_predicted - y_true))
     bias_d = np.mean(y_predicted - y_true)
     w1 = w1 - rate * w1d
     w2 = w2 - rate * w2d
     w3 = w3 - rate * w3d
     bias = bias - rate * bias_d

     print(f"Epoch {i + 1}: w1={w1:.4f}, w2={w2:.4f}, w3={w3:.4f}, bias={bias:.4f}, loss={loss:.4f}")


     if loss <= loss_threshold:
      break
  return w1,w2,bias
```
