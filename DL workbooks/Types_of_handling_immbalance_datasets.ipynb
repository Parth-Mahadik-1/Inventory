#build an ANN model

```python
def ANN(x_train,x_test,y_train,y_test,loss,weights):
  model = keras.Sequential([
      keras.layers.Dense(26,input_shape=(26,),activation='relu'),
      keras.layers.Dense(15,activation='relu'),
      keras.layers.Dense(1,activation='sigmoid'),
  ])
  model.compile(
      optimizer='adam',
      loss=loss,
      metrics=['accuracy']
  )
  '''
  CLASS weight explination:

  If we have a dataset where:

Class 0 has 90% of the data
Class 1 has only 10% of the data
Then, we can set weights = {0: 1.0, 1: 9.0} to give class 1 more weight.

If weights == -1, the model learns normally, potentially biased toward class 0.
If weights is given, it adjusts the loss contribution, making the model more balanced.'''

  if weights == -1:
    model.fit(x_train,y_train,epochs=50)
  else:
    model.fit(x_train,y_train,epochs=50,class_weight=weights)

  print(model.evaluate(x_test,y_test))
  yp = model.predict(x_test)
  yp=np.round(yp)

  print(f"classification report {classification_report(y_test,yp)}")

  return yp
```

```python
ANN(x_train,x_test,y_train,y_test,'binary_crossentropy',-1)
```

1)Method 1 - undersampling <br>Scenario: Detecting Fraudulent Transactions
Consider a dataset where:<br>

Class 0 (Legitimate transactions): 95,000 samples<br>
Class 1 (Fraudulent transactions): 5,000 samples<br><br>
This dataset is highly imbalanced, and a model trained on it may ignore fraudulent transactions because they appear much less frequently.<br>

Applying Undersampling
Instead of using all 95,000 legitimate transactions, we randomly select only 5,000 legitimate transactions.<br><br>
Now, we have:<br><br>
Class 0 (Legitimate transactions): 5,000 samples<br>
Class 1 (Fraudulent transactions): 5,000 samples<br><br>
This balances the dataset and ensures the model learns equally from both classes.

```python
#get count of samples
count_class_0 , count_class_1 = df3.Churn.value_counts()
```

```python
df_class_0 = df3[df3['Churn']==0]
df_class_1 = df3[df3['Churn']==1]
```

```python
df_class_0_under_sample = df_class_0.sample(count_class_1)
df_under_final = pd.concat([df_class_0_under_sample,df_class_1],axis=0)
print('random sample to be: ')
print(df_under_final.Churn.value_counts())
```

```python
x = df_under_final.drop('Churn',axis='columns')
y = df_under_final.Churn
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=5,stratify = y)
```

```python
y_train.value_counts()
```

```python
y_test.value_counts()
```

```python
ANN(x_train,x_test,y_train,y_test,'binary_crossentropy',-1)
```

#Method2: Oversampling<br>
Scenario: Detecting Fraudulent Transactions<br>
Consider a dataset where:<br>

Class 0 (Legitimate transactions): 95,000 samples<br>
Class 1 (Fraudulent transactions): 5,000 samples<br><br>
This dataset is highly imbalanced, meaning the model may learn to predict most transactions as legitimate, ignoring fraudulent cases due to their lower frequency.<br><br>

Applying Oversampling:<br>
Instead of reducing the majority class size, we increase the number of fraudulent transactions by duplicating or generating synthetic samples using oversampling techniques like Random Oversampling or SMOTE.<br><br>

Now, we have:<br><br>

Class 0 (Legitimate transactions): 95,000 samples<br>
Class 1 (Fraudulent transactions): 95,000 samples<br>
This balances the dataset and ensures the model learns equally from both classes. Oversampling prevents data loss and improves fraud detection accuracy.<br>

```python
df_class_1_over_sample = df_class_1.sample(count_class_0,replace=True)
df_under_final_over = pd.concat([df_class_0,df_class_1_over_sample],axis=0)
print('random sample to be: ')
print(df_under_final_over.Churn.value_counts())
```

```python
x = df_under_final_over.drop('Churn',axis = 'columns')
y = df_under_final_over.Churn
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=5,stratify = y)
```

```python
y_train.value_counts()
```

```python
y_test.value_counts()
```

```python
ANN(x_train,x_test,y_train,y_test,'binary_crossentropy',-1)
```

Method - SMOTE
Why Use SMOTE?<br>
Increases the number of minority class samples without overfitting (which happens in random oversampling).<br>
Helps the model learn better by creating more diverse data points.<br>
Improves classification performance, especially recall for the minority class.<br><br>
âœ… Pros:<br>

1)Avoids overfitting (unlike simple duplication).<br>
2)Works well with low-dimensional data.<br>
3)Improves recall for the minority class.<br>
âŒ Cons:<br><br>

1)an introduce noise if synthetic samples are not meaningful.<br>
2)Not effective for high-dimensional datasets (as distances may become less meaningful).<br>
3)Doesn't work well when minority class instances are highly overlapping with the majority class.<br>


```python
from imblearn.over_sampling import SMOTE
x = df3.drop('Churn',axis='columns')
y = df3.Churn
smote = SMOTE(sampling_strategy='minority')
x_sm,y_sm = smote.fit_resample(x,y)
```

```python
y_sm.value_counts()
```

```python
x_train,x_test,y_train,y_test = train_test_split(x_sm,y_sm,test_size=0.2,random_state=5,stratify = y_sm)
```

```python
ANN(x_train,x_test,y_train,y_test,'binary_crossentropy',-1)
```

### **Conclusion**  
We observed a significant improvement in both the **F1 score** and overall model performance when the dataset was balanced. Proper handling of class imbalance ensures that the model **does not favor the majority class**, leading to more accurate and fair predictions.  

Thus, maintaining a **balanced dataset** plays a crucial role in enhancing the **accuracy and reliability** of the model.
